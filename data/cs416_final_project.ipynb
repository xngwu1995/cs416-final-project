```python
import json
  
# Opening JSON file
f = open('data.json',)
  
# returns JSON object as 
# a dictionary
data = json.load(f)
  
# Iterating through the json
# list
ans_1800 = {'population': {'europe': 0, 'americas': 0, 'asia': 0, 'africa': 0},
            'income': {'europe': 0, 'americas': 0, 'asia': 0, 'africa': 0}}
ans_2014 = {'population': {'europe': 0, 'americas': 0, 'asia': 0, 'africa': 0},
            'income': {'europe': 0, 'americas': 0, 'asia': 0, 'africa': 0}}
for val in data[0]['countries']:
    ans_1800['population'][val['continent']] += int(val['population'])
    if val['income']:
        ans_1800['income'][val['continent']] += int(val['income']) * int(val['population'])

for val in data[-1]['countries']:
    if val['population']:
        ans_2014['population'][val['continent']] += int(val['population'])
    else:
        val['population'] = 0
    if val['income']:
        ans_2014['income'][val['continent']] += int(val['income']) * int(val['population'])
```

```python
print(ans_1800)
print(ans_2014)
```

```python
with open('1800_summary.json', 'w') as outfile:
    json.dump(ans_1800, outfile)
with open('2014_summary.json', 'w') as outfile:
    json.dump(ans_2014, outfile)
47576633919663
24163794777957
27223554647917
```

```python
import collections
ans = {'China': [], 'United States': []}
for i in range(len(data)):
    for j in range(len(data[i]['countries'])):
        if data[i]['countries'][j]['country'] == 'China':
            temp = {}
            val = data[i]['countries'][j]
            temp['Population'] = val['population']
            temp['Income'] = val['income']
            temp['Life_Exp'] = val['life_exp']
            temp['Year'] = data[i]['year']
            ans['China'].append(temp)
        if data[i]['countries'][j]['country'] == 'United States':
            temp = {}
            val = data[i]['countries'][j]
            temp['Population'] = val['population']
            temp['Income'] = val['income']
            temp['Life_Exp'] = val['life_exp']
            temp['Year'] = data[i]['year']
            ans['United States'].append(temp)
with open('China vs US.json', 'w') as outfile:
    json.dump(ans, outfile)
# print(ans)
```
